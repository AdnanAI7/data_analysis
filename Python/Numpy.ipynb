## Array Creation
There are several ways to create arrays

```python
import numpy as np 
a = np.array([2,3,4])
print(a)
```

```python
a.dtype
```

```python
b = np.array([1.2, 3.5, 5.1])
b.dtype
```

```python
a = np.array(1,2,3,4) #Wrong
a = np.array([1,2,3,4]) #Correct
```

`array` transformers sequences of sequences into two dimensional arrays, sequences of sequences of sequences into three dimensional arrays, and so on

```python
b = np.array([(1.5,2,3), (4,5,6)])
print(b)
```

The type of the array can also be explicitly specified at creation time:

```python
c = np.array([[1,2],[3,4]], dtype=complex)
print(c)
```

* Often the elements of an array are originally unknown, but it's size is known. Hence, Numpy offers several functions to create arrays with initial placeholder content. These minimize the necessity of growing arrays, an expensive operation.
* The function `zeros` creates an array full of zeros, the function `ones` creates an array full of ones, and the function `empty` creates an array whose initial content is random and depends on the state of the memory. By default, the dtype of the created array is `float64`.

```python
np.zeros((3,4))
```

```python
np.ones((2,3,4), dtype=np.int16)
```

```python
np.empty((2,3))
```

To create sequence of numbers, NumPy provides a function analogous to `range` that returns arrays instead of lists.

```python
np.arange(10,30,5)
```

```python
np.arange(0, 2, 0.3)
```

When `arange` is used with floating point arguments, it is generally not possible to predict the number of elements obtained, due to the finite floating point precision. Fot this reason, it is usually better to use the function `linspace` that receives as an argument the numbe of elements that we want, instead of the step:

```python
from numpy import pi
np.linspace(0,2,9)
```

```python
x = np.linspace(0, 2*pi, 100)
f = np.sin(x)
```

## Printing Arrays

When you print an array. Numpy displays it in a similar way to nested lists, but with the following layout:
- The last axis is printed from left to right, 
- The second-to-last is printed from top to bottom, 
- The rest are also printed from top to bottom, with each slice separated from the next bu an empty line. 

**One-Dimensional arrays are then printed as rows, bidimensionals are matrices and tridimensionals as lists of matrices.**

```python
a = np.arange(6)
print(a)
```

```python
b = np.arange(12).reshape(4,3)
print(b)
```

```python
c = np.arange(24).reshape(2,3,4)
print(c)
```

If an array is too large to be printed, NumPy automatically skips the central part of the array and only prints the corners:

```python
print(np.arange(10000))
```

```python
sys = np.arange(10000).reshape(100,100)
print(sys)
```

To disable this behaviour and force NumPy to print the entire array. you can change the printing options using `set_printoptions`.

```python
import sys

np.set_printoptions(threshold=sys.maxsize)
```

# Data Analysis with Histograms and Box Plots

This notebook demonstrates how to analyze randomly generated data using histograms and box plots for different subsamples of data.

## Import Required Libraries

First, we need to import NumPy for data manipulation and Matplotlib for visualization.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the style for better visualizations
plt.style.use('seaborn-v0_8-whitegrid')
```

## Generate Random Data

We'll create four distinct subsamples of data:
1. Spread - random values across a range
2. Center - fixed values at the midpoint
3. Flier_high - high outliers
4. Flier_low - low outliers

```python
# Fix random seed for reproducibility
np.random.seed(42)

# Generate the four subsamples of data
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100

# Combine all data into one array
all_data = np.concatenate((spread, center, flier_high, flier_low))

# Create a dictionary to hold our datasets
datasets = {
    'Spread': spread,
    'Center': center, 
    'High Outliers': flier_high,
    'Low Outliers': flier_low,
    'All Data': all_data
}

# Print summary statistics
for name, data in datasets.items():
    print(f"{name}: min={data.min():.1f}, max={data.max():.1f}, mean={data.mean():.1f}, std={data.std():.1f}")
```

## Create Individual Histograms

Let's create separate histograms for each subsample while keeping the x-axis range consistent across all plots for better comparison.

```python
# Define common x-axis limits
x_min = min(flier_low.min(), spread.min(), center.min(), flier_high.min())
x_max = max(flier_low.max(), spread.max(), center.max(), flier_high.max())
bin_count = 30

plt.figure(figsize=(14, 8))

# Create individual histograms for each dataset
for i, (name, data) in enumerate(list(datasets.items())[:-1], 1):  # Exclude "All Data" for now
    plt.subplot(2, 2, i)
    plt.hist(data, bins=bin_count, alpha=0.7, color=f'C{i-1}')
    plt.title(f'Histogram of {name}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.xlim(x_min, x_max)  # Keep x-axis consistent

plt.tight_layout()
plt.show()
```

## Plot Combined Histogram

Now let's create a combined histogram showing all the data together, and then another visualization showing all subsamples in a single figure with multiple histograms for direct comparison.

```python
# Combined histogram of all data
plt.figure(figsize=(10, 6))
plt.hist(all_data, bins=bin_count, alpha=0.7, color='steelblue')
plt.title('Histogram of All Data Combined')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()

# Overlay of all datasets
plt.figure(figsize=(12, 6))
for i, (name, data) in enumerate(list(datasets.items())[:-1]):  # Exclude "All Data"
    plt.hist(data, bins=bin_count, alpha=0.5, label=name)
plt.title('Comparison of All Subsamples')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(x_min, x_max)
plt.show()
```

## Create Individual Box Plots

Box plots are great for visualizing the distribution of data, especially for identifying outliers and comparing distributions across groups.

```python
plt.figure(figsize=(14, 8))

# Common y-axis limits
y_min = min(flier_low.min(), spread.min(), center.min(), flier_high.min())
y_max = max(flier_low.max(), spread.max(), center.max(), flier_high.max())

# Create individual box plots
for i, (name, data) in enumerate(list(datasets.items())[:-1], 1):  # Exclude "All Data" for individual plots
    plt.subplot(2, 2, i)
    plt.boxplot(data)
    plt.title(f'Box Plot of {name}')
    plt.ylabel('Value')
    plt.ylim(y_min, y_max)  # Keep y-axis consistent
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Create Combined Box Plot

Now let's create a combined box plot showing all subsamples together for direct comparison.

```python
# Create a list of data arrays excluding "All Data"
data_list = [spread, center, flier_high, flier_low]
labels = ['Spread', 'Center', 'High Outliers', 'Low Outliers']

plt.figure(figsize=(10, 6))
box_plot = plt.boxplot(data_list, patch_artist=True, labels=labels)

# Customize box colors
colors = ['lightblue', 'lightgreen', 'pink', 'lightyellow']
for patch, color in zip(box_plot['boxes'], colors):
    patch.set_facecolor(color)

plt.title('Comparison of Data Distributions')
plt.ylabel('Value')
plt.grid(True, alpha=0.3)
plt.show()
```

## Create Violin Plots for Additional Insight

Violin plots combine box plots with kernel density estimates, providing more detail about the data distribution.

```python
plt.figure(figsize=(12, 6))
plt.violinplot(data_list, showmeans=True, showmedians=True)

# Add labels
plt.xticks(np.arange(1, len(labels) + 1), labels)
plt.title('Violin Plots of All Subsamples')
plt.ylabel('Value')
plt.grid(True, alpha=0.3)
plt.show()
```

## Compare Visualizations

Let's analyze what the different visualization methods reveal about the data distribution in each subsample.

```python
# Create a figure with both box plots and histograms
fig, axes = plt.subplots(2, len(labels), figsize=(16, 8), sharex='col')
fig.suptitle('Comparison of Box Plots and Histograms', size=16)

# Loop through each dataset
for i, (name, data) in enumerate(zip(labels, data_list)):
    # Box plot in the top row
    axes[0, i].boxplot(data)
    axes[0, i].set_title(name)
    if i == 0:
        axes[0, i].set_ylabel('Box Plot')
    
    # Histogram in the bottom row
    axes[1, i].hist(data, bins=15, orientation='horizontal')
    axes[1, i].set_xlabel('Frequency')
    if i == 0:
        axes[1, i].set_ylabel('Histogram')
    
    # Set common y limits for better comparison
    axes[0, i].set_ylim([y_min, y_max])
    axes[1, i].set_ylim([y_min, y_max])

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

## Key Observations

From our visualizations, we can observe:

1. **Spread data**: Shows a uniform distribution across the 0-100 range
2. **Center data**: Shows a very tight distribution centered at 50
3. **High Outliers**: All values are above 100, clearly separated from the other groups
4. **Low Outliers**: All values are below 0, forming a distinct group

The box plots effectively show the median, quartiles, and outliers, while histograms give us a better sense of the frequency distribution. Violin plots combine these approaches to show both distribution shape and summary statistics.

Different visualization methods have different strengths:
- Histograms are best for showing the frequency distribution and identifying modes
- Box plots excel at showing the five-number summary and highlighting outliers
- Violin plots combine features of both to show distribution shape along with key statistics

```python
# Calculate and display specific statistics about each dataset
plt.figure(figsize=(10, 6))

# Bar chart of standard deviations
std_devs = [datasets[name].std() for name in labels]
plt.bar(labels, std_devs, alpha=0.7)
plt.title('Standard Deviation Comparison')
plt.ylabel('Standard Deviation')
plt.grid(True, alpha=0.3)
plt.show()
```
